```python
import pandas as pd
#TN_01
tn01_col = pd.read_csv('TN_01_merged_data_summary.csv', usecols=[0, 1])
print(tn01_col.info())

#TN_16
tn16_col = pd.read_csv('TN_16_merged_data_summary.csv', usecols=[0, 1])
#rename the column 'Unnamed: 0'
tn16_col.rename(columns={'Unnamed: 0': 'header'}, inplace=True)
print(tn16_col.info())

#TN_19
tn19_col = pd.read_csv('TN_19_merged_data_summary.csv', usecols=[0, 1])
tn19_col.rename(columns={'Unnamed: 0': 'header'}, inplace=True)
print(tn19_col.info())

#TN_20
tn20_col = pd.read_csv('TN_20_merged_data_summary.csv', usecols=[0, 1])
tn20_col.rename(columns={'Unnamed: 0': 'header'}, inplace=True)
print(tn20_col.info())
```

```python
df_tn20_99 = pd.read_csv('TN_20_merged_data.csv', usecols=tn_20_col_99, delimiter=',', quotechar='"', on_bad_lines='warn')
df_tn20_99 = df_tn20_99.astype('category')
df_tn20_99 = df_tn20_99.drop_duplicates()
df_tn20_99.info()
```

```python
import pandas as pd

# Identify the common headers
common_headers_01_16 = set(df_tn01_99.columns).intersection(df_tn16_99.columns)
print('count of common headers: ', len(common_headers_01_16))

# Step 1: Ensure shared categorical columns have the same categories
for col in common_headers_01_16:
    if isinstance(df_tn01_99[col].dtype, pd.CategoricalDtype) and isinstance(df_tn16_99[col].dtype, pd.CategoricalDtype):
        # Create a union of categories
        categories = set(df_tn01_99[col].cat.categories).union(set(df_tn16_99[col].cat.categories))
        
        # Update the categories for both columns
        df_tn01_99[col] = df_tn01_99[col].cat.set_categories(categories)
        df_tn16_99[col] = df_tn16_99[col].cat.set_categories(categories)

# Step 2: Convert non-matching categorical columns to 'object' type
for col in common_headers_01_16:
    if df_tn01_99[col].dtype != df_tn16_99[col].dtype:
        df_tn01_99[col] = df_tn01_99[col].astype('object')
        df_tn16_99[col] = df_tn16_99[col].astype('object')

# Merge the DataFrames using Pandas
df_tn01_tn16_merge = pd.merge(df_tn01_99, df_tn16_99, on=list(common_headers_01_16), how='outer', suffixes=('', '_dup_tn16'))

# Print DataFrame information and the first few rows
print(df_tn01_tn16_merge.info())
df_tn01_tn16_merge.head()
```

```python
# Function to check if a value is not an integer
def is_not_integer(val):
    try:
        float_val = float(val)  # Convert to float first
        int_val = int(float_val)
        return float_val != int_val  # Check if float value is not equal to its integer conversion
    except ValueError:
        return True  # Non-numeric values are also considered as not integer

# Apply the function to filter out non-integer values
MaskID_non_integer_values = df_tn01_tn16_tn19_tn20_merge['MaskID'][df_tn01_tn16_tn19_tn20_merge['MaskID'].apply(is_not_integer)]
print(MaskID_non_integer_values)
```

```python
# List of columns to convert from object to category
columns_to_convert = ['HBsAb', 'ConMedsFrequencyOtherDetail', 'Height', 'Weight', 'ConMedsIntervalOtherDetails', 
                      'TestingPlaceForAutoAntibodyOut', 'SelectStudyForPreventionOfT1D', 'ScoringQ5AndQ6']

for col in columns_to_convert:
    df_tn01_tn16_tn19_tn20_merge[col] = df_tn01_tn16_tn19_tn20_merge[col].astype('category')
```

```python
import pandas as pd
import numpy as np

# Function to check if all non-NaN values in a column are integers
def all_integers(column):
    return column.dropna().apply(float.is_integer).all()

# Iterate through each column in the DataFrame
for column in df_tn01_tn16_tn19_tn20_merge.columns:
    # Check if the column is of object type and convert 'nan' to np.nan
    if df_tn01_tn16_tn19_tn20_merge[column].dtype == 'object':
        df_tn01_tn16_tn19_tn20_merge[column] = df_tn01_tn16_tn19_tn20_merge[column].replace('nan', np.nan)

    # Convert object types to numeric (float or int as appropriate)
    if pd.api.types.is_object_dtype(df_tn01_tn16_tn19_tn20_merge[column]):
        df_tn01_tn16_tn19_tn20_merge[column] = pd.to_numeric(df_tn01_tn16_tn19_tn20_merge[column], errors='coerce')
    
    # Check if the column is float
    if pd.api.types.is_float_dtype(df_tn01_tn16_tn19_tn20_merge[column]):
        # If all non-NaN values are integers, convert to 'Int64'
        if all_integers(df_tn01_tn16_tn19_tn20_merge[column]):
            df_tn01_tn16_tn19_tn20_merge[column] = df_tn01_tn16_tn19_tn20_merge[column].astype('Int64')
        else:
            # Otherwise, check if it is safe to convert to 'float16'
            max_val = df_tn01_tn16_tn19_tn20_merge[column].max()
            min_val = df_tn01_tn16_tn19_tn20_merge[column].min()
            if max_val < np.finfo(np.float16).max and min_val > np.finfo(np.float16).min:
                df_tn01_tn16_tn19_tn20_merge[column] = df_tn01_tn16_tn19_tn20_merge[column].astype('float16')
```

```python
for col in columns_to_convert:
    df_tn01_tn16_tn19_tn20_merge[col].replace('nan', np.nan, inplace=True)
```

```python
import pandas as pd

# Calculate the percentage of missing values
missing_percentage = df_tn01_tn16_tn19_tn20_merge.isnull().mean() * 100

# Calculate the memory usage for each column
memory_usage = df_tn01_tn16_tn19_tn20_merge.memory_usage(deep=True)

# Initialize a list to hold the data for the new DataFrame
data = []

# Iterate through each column in the original DataFrame
for column in df_tn01_tn16_tn19_tn20_merge.columns:
    # Find the top 5 most frequent values (excluding NaN) and their counts
    top_5_values_counts = df_tn01_tn16_tn19_tn20_merge[column].value_counts(dropna=True).head(5)
    top_5_values = top_5_values_counts.index.tolist()
    top_5_counts = top_5_values_counts.tolist()

    # Ensure the lists have 5 elements by padding with None if necessary
    top_5_values += [None] * (5 - len(top_5_values))
    top_5_counts += [None] * (5 - len(top_5_counts))

    # Determine the data type of the column
    column_dtype = str(df_tn01_tn16_tn19_tn20_merge[column].dtype)

    # Get memory usage for the column
    column_memory_usage = memory_usage[column]

    # Append the data for this column to the list
    data.append([missing_percentage[column], column_dtype, column_memory_usage] + top_5_values + top_5_counts)

# Define the columns for the new DataFrame
columns = ['Missing Percentage', 'Value Type', 'Memory Usage (Bytes)',
           'Top 1', 'Top 2', 'Top 3', 'Top 4', 'Top 5',
           'Top 1 Value Count', 'Top 2 Value Count', 'Top 3 Value Count', 
           'Top 4 Value Count', 'Top 5 Value Count']

# Create the new DataFrame
summary_df = pd.DataFrame(data, index=df_tn01_tn16_tn19_tn20_merge.columns, columns=columns)

# Sort the DataFrame by 'Missing Percentage' in descending order
summary_df = summary_df.sort_values(by='Missing Percentage', ascending=False)

# Save the DataFrame to a CSV file
summary_df.to_csv('TN_all_summary_99_2.csv')
```
