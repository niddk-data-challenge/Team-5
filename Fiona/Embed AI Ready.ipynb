```python
import pandas as pd

merged_df = pd.read_csv('Processing data/TN/TN_allinone0112.csv')
merged_df.info()
```

```python
#identify categorical columns and numeric columns first 

categorical_columns = merged_df.select_dtypes(include=['object', 'category']).columns
numerical_columns = merged_df.select_dtypes(include=['int', 'float']).columns
print("Categorical columns:", categorical_columns)
print('number of categorical columns:',len(categorical_columns))
print("Numerical columns:", numerical_columns) 
print('number of numerical columns:',len(numerical_columns))
```

```python
numerical_columns
```

```python
missing=[]
for i in numerical_columns:
    count_null=merged_df[i].isnull().sum()
    print(i)
    #print(count_null)
    all1=len(merged_df[i])
    #print(all)
    null_per=count_null/all1
    print(null_per*100,'%')
    if null_per*100<50:
        missing.append(i)
```

```python
missing_above_50=[]
for i in numerical_columns:
    count_null=merged_df[i].isnull().sum()
    print(i)
    #print(count_null)
    all1=len(merged_df[i])
    #print(all)
    null_per=count_null/all1
    print(null_per*100,'%')
    if null_per*100>=50 and null_per*100<99:
        missing_above_50.append(i)
```

```python
missing=missing[1:]
```

```python
df = merged_df.drop(merged_df.columns[0], axis=1)
```

```python
df[missing]
```

```python
from sklearn.impute import KNNImputer
import pandas as pd
# Filter the DataFrame to include only the selected columns
df_to_impute = df[missing]

# Initialize the KNNImputer with the desired number of neighbors (k)
knn_imputer = KNNImputer(n_neighbors=5)  

# Perform K-NN imputation on the selected columns
df_imputed = knn_imputer.fit_transform(df_to_impute)

# The result will be a NumPy array; convert it back to a DataFrame with column names
df_imputed = pd.DataFrame(df_imputed, columns=df_to_impute.columns)

# Merge the imputed columns back into the original DataFrame
df[missing] = df_imputed

# Display the DataFrame with missing values imputed for selected columns
print(df)
```
