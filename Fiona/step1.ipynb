```python
import pandas as pd
import os
import numpy as np
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
merged_df_info=pd.read_csv('Processing data/TN/TN01_info.csv')
merged_df16_info=pd.read_csv('Processing data/TN/TN16_info.csv')
merged_df19_info=pd.read_csv('Processing data/TN/TN19_info.csv')
merged_df20_info=pd.read_csv('Processing data/TN/TN20_info.csv')
```

```python
merged_df_info = merged_df_info.dropna(axis=1, thresh=100)
merged_df16_info = merged_df16_info.dropna(axis=1, thresh=100)
merged_df19_info = merged_df19_info.dropna(axis=1, thresh=100)
merged_df20_info = merged_df20_info.dropna(axis=1, thresh=100)
```

```python
merged_df16_info = pd.concat([merged_df16_info, merged_df19_info], ignore_index=True) 
merged_df16_info = pd.concat([merged_df16_info, merged_df20_info], ignore_index=True) 
merged_df_info = pd.concat([merged_df_info, merged_df16_info], ignore_index=True) 
```

```python
vd=['mg','Rescreening #1',".:",'Baseline 0B','Day 2',"Initial Visit",'Baseline 0A','Baseline','PRN',
                '2 weeks','6 weeks','4 weeks', '10 weeks','8 weeks','24 weeks','12 weeks','48 weeks','36 weeks','14 weeks','20 weeks','16 weeks','18 weeks','22 weeks','96 weeks','72 weeks']
merged_df_info = merged_df_info[~merged_df_info['Visit'].isin(vd)]
```

```python
#merged_df_info.shape
```

```python
#len(merged_df_info['MaskID'].unique())
```

```python
#patients should do only do one type of test at one time. If there are duplicate ones, drop the row with less data
row_non_null_counts = merged_df_info.notnull().sum(axis=1)

# Add the non-null count as a new column to the DataFrame
merged_df_info['Non_Null_Count'] = row_non_null_counts

# Sort the DataFrame by Non_Null_Count in descending order
sorted_df = merged_df_info.sort_values(by='Non_Null_Count', ascending=False)

# Group by both 'MaskID' and 'Visit'
grouped_df = sorted_df.groupby(['MaskID', 'Visit'])

# Take the first row from each group
result_df = grouped_df.first()

# Drop the 'Non_Null_Count' column if you no longer need it
result_df = result_df.drop(columns=['Non_Null_Count'])

# Display the result DataFrame
print(result_df)
```

```python
#result_df.shape
```

```python
#drop the rows and columns with all NULL in it.
result_df.dropna(axis=1, how='all', inplace=True)
result_df.dropna(axis=0, how='all', inplace=True)
df=result_df
```

Handle wrong data type

```python
#identify categorical columns and numeric columns first 

categorical_columns = df.select_dtypes(include=['object', 'category']).columns
numerical_columns = df.select_dtypes(include=['int', 'float']).columns

print("Categorical columns:", categorical_columns)
print('number of categorical columns:',len(categorical_columns))
print("Numerical columns:", numerical_columns) 
print('number of numerical columns:',len(numerical_columns))
```

```python
#1st handle wrong datatype
#for any categorical variable, if it contains more than 10 unique values, then we can use domain knowledge to judge 
# if it is better to convert it to 'int' datatype.
for i in categorical_columns:
    if len(df[i].unique())>20:
        print(i)
        print(df[i].unique())
```

```python
#these are the categorical_columns that need to be converted to numeric
to_convert=['BasophilsResultsOR','bloodpressureDiastolic','TimeDoseGivenMin','StudyDrugReturnedDay','CapsulesReturned',
            'StudyDrugDispensedDay','CapsulesDispensed','ParticipantDateDispensedDay1_1','ParticipantDateDispensedDay2_1',
            'ParticipantDateDispensedDay3_1','ParticipantDateDispensedDay4_1',
    'BasophilsResultsOR','bloodpressureDiastolic','TimeDoseGivenMin','StudyDrugReturnedDay','CapsulesReturned',
            'StudyDrugDispensedDay','CapsulesDispensed',
    'TimeOfCollectionHH',
            'TimeOfCollectionMM',
            'TimeResultsProcessedByLabHH',
            'TimeResultsProcessedByLabMM',
            'Rbcresults',
            'HemoglobinResults',
            'HematocritResults',
            'MCVresults',
            'PlateletCountResultsNormal',
            'MCHresults',
            'MCHCresults',
            'WBCresults',
            'NeutrophilsResults',
            'NeutrophilsResultsOR',
            'LymphocytesResults',
            'LymphocytesResultsOR',
            'MonocytesResults',
            'MonocytesResultsOR',
            'EosinophilsResults',
            'EosinophilsResultsOR',
            'BasophilsResults',
            'Weight',
            'Height',
            'bloodpressureSystolic',
            'PreDoseGlucose',
    'Rbcresults',
            'MCHCresults',
            'MonocytesResultsOR',
           'HemoglobinResults',
            'HematocritResults',
            'MCVresults',
            'PlateletCountResultsNormal',
            'MCHresults',
            'WBCresults',
            'NeutrophilsResults',
            'NeutrophilsResultsOR',
            'LymphocytesResults',
            'LymphocytesResultsOR',
            'MonocytesResults',
            'EosinophilsResults',
            'EosinophilsResultsOR'
           ]
```

```python
#len(to_convert)
```

```python
#convert some invalid values in the columns that can only contain number to "NaN" values
for i in to_convert:
    df[i] = df[i].replace(['Measured','.','Self-reported','<LLOD','> 1000'], np.nan)
    print(i)
    print(df[i].dtypes)
    df[i] = df[i].astype(float)
```

```python
missing=[]
for i in numerical_columns:
    count_null=df[i].isnull().sum()
    print(i)
    #print(count_null)
    all1=len(df[i])
    #print(all)
    null_per=count_null/all1
    print(null_per*100,'%')
    if null_per*100<60:
        missing.append(i)
```

```python
#for the column with missing value less than 50% and unique values <20, we consider them as categorical value, so convert
#their data type to 'object'
for i in missing:
    #print(i)
    #print(df[i].unique())
    if len(df[i].unique())<20:
        print(i)
        print(df[i].unique())
        df[i] = df[i].astype('object')
```

```python
df.dropna(axis=1, how='all', inplace=True)
df.dropna(axis=0, how='all', inplace=True)
```
