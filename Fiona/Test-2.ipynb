```python
# Directory containing the CSV files
#directory = ".././studies/TN_01/TrialNet_01_Data"

def countheaders(dataset_index,exttype):

    directory = "/home/ec2-user/SageMaker/studies/TN_{0}/TrialNet_{0}_Data".format(str(dataset_index))

    # List to store (file name, column header) tuples
    headers_list = []

    # Iterate through each CSV file in the directory
    for file in os.listdir(directory):
        if file.endswith(exttype):
            file_path = os.path.join(directory, file)
            try:
                # Read headers from the CSV file
                with open(file_path, 'r') as f:
                    df = pd.read_csv(file_path)
                    headers_list=headers_list+df.columns.tolist()
            except:
                print("skip invalid csv file:"+file_path)

    # Count the occurrences of each value in the list
    value_counts = Counter(headers_list)

    # Print the counts
    print("file number:"+str(len(file)))

    for value, count in value_counts.items():
        print(f"{value}: {count}")

    return headers_list, value_counts
```

```python
def mergecvs(datasetindex,dt):

    # Directory containing the CSV files
    #directory = ".././studies/TN_01/TrialNet_01_Data"

    directory = "/home/ec2-user/SageMaker/studies/TN_{0}/TrialNet_{0}_Data".format(str(datasetindex))

    merged_df=pd.DataFrame()

    # Iterate through each CSV file in the directory
    for file in os.listdir(directory):
        if file.endswith(dt):
            file_path = os.path.join(directory, file)
        try:
            
            df = pd.read_csv(file_path)  # Read each CSV file into a DataFrame
            merged_df = pd.concat([merged_df, df], ignore_index=True)  # Concatenate DataFrames
        except:
            print("invald csv file:"+file_path)

    # Custom aggregation function to keep the first non-null value
#    def first_not_null(series):
#        for value in series:
#            if not pd.isna(value):
#                return value
#    return None

    # Group by 'group_col' and apply the custom aggregation function
#    merged_df = merged_df.groupby('MaskID').agg(first_not_null).reset_index()

    # Save the merged DataFrame to a new CSV file
    merged_df.to_csv('merged_output_hu_{0}.csv'.format(str(datasetindex)), index=False)
    
    return merged_df
```

```python
import pandas as pd

data = {'1': ['A', 'A', 'B', 'B', 'C','D'],
        '2': [1, 2, None, 4, None,None],
        '3': [None, 2, 3, 4, 5,None]}
df = pd.DataFrame(data)

# Define a custom aggregation function to keep the first not-null value
def first_not_null(series):
    for value in series:
        if not pd.isna(value):
            return value
    return None

# Group by '1' and apply the custom aggregation function to each column
result = df.groupby('1').agg(first_not_null).reset_index()

print(result)
```

```python
import pandas as pd
import numpy as np

# Sample DataFrame
data = {'group_col': ['A', 'A', 'B', 'B', 'C'],
        'value_col': [1, 2, None, 4, 5]}
df = pd.DataFrame(data)

# Custom aggregation function to keep the first non-null value
def first_not_null(series):
    for value in series:
        if not pd.isna(value):
            return value
    return np.nan

# Group by 'group_col' and apply the custom aggregation function
result = df.groupby('group_col')['value_col'].agg(first_not_null).reset_index()

print(result)
```
