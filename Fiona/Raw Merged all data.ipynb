```python
import pandas as pd
import os
```

```python
#load all data
df01=pd.read_csv('merged_output_01.csv')
#print(df01.shape)
df16=pd.read_csv('merged_output_16.csv')
print(df16.shape)
df19=pd.read_csv('merged_output_19.csv')
print(df19.shape)
df20=pd.read_csv('merged_output_20.csv')
print(df20.shape)
```

```python
#replace TNXX column name as TNXX_ index, save the list into TNXX column names

df16_column=df16.columns.tolist()
new_column_names = ["TN16_" + str(i) for i in range(len(df16_column))]
df16.columns = new_column_names
df16.rename(columns={df16.columns[0]: 'MaskID'}, inplace=True)
file_path="TN16 column names"
with open(file_path, 'w') as file:
    for item in df16_column:
        file.write(f"{item}\n")
print(f"List saved to {file_path}")


df01_column=df01.columns.tolist()
new_column_names = ["TN01_" + str(i) for i in range(len(df01_column))]
df01.columns = new_column_names
df01.rename(columns={df01.columns[0]: 'MaskID'}, inplace=True)
file_path="TN01 column names"
with open(file_path, 'w') as file:
    for item in df01_column:
        file.write(f"{item}\n")
print(f"List saved to {file_path}")


df19_column=df19.columns.tolist()
new_column_names = ["TN19_" + str(i) for i in range(len(df19_column))]
df19.columns = new_column_names
df19.rename(columns={df19.columns[0]: 'MaskID'}, inplace=True)
file_path="TN19 column names"
with open(file_path, 'w') as file:
    for item in df19_column:
        file.write(f"{item}\n")
print(f"List saved to {file_path}")


df20_column=df20.columns.tolist()
# Replace the column names with "TN01_" + index
new_column_names = ["TN20_" + str(i) for i in range(len(df20_column))]
df20.columns = new_column_names
df20.rename(columns={df20.columns[0]: 'MaskID'}, inplace=True)
file_path="TN20 column names"
with open(file_path, 'w') as file:
    for item in df20_column:
        file.write(f"{item}\n")
print(f"List saved to {file_path}")
```

```python
print(df01['MaskID'].dtype)
print(df16['MaskID'].dtype)
print(df19['MaskID'].dtype)
print(df20['MaskID'].dtype)
```

```python
# Attempt to convert 'MaskID' to int64, drop rows with conversion errors, and print them
try:
    df16['MaskID'] = df16['MaskID'].astype('int64')
except ValueError as e:
    error_indices = df16.index[df16['MaskID'].apply(lambda x: not x.isdigit())]
    error_rows = df16.loc[error_indices]
    print("Rows with 'MaskID' conversion errors:")
    print(error_rows)
    df16.drop(error_indices, inplace=True)
    
print(df01['MaskID'].dtype)
print(df16['MaskID'].dtype)
print(df19['MaskID'].dtype)
print(df20['MaskID'].dtype)
```

```python
# Merge the DataFrames sequentially with 'MaskID' as the primary key
merged_df = df16.merge(df19, on='MaskID', how='outer') \
    .merge(df20, on='MaskID', how='outer') \
    .merge(df01, on='MaskID', how='outer')
```

```python
merged_df.to_csv('merged_df.csv',index=False)
```
