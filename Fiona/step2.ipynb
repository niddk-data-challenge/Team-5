```python
import pandas as pd
import os
```

```python
info=pd.read_csv('Processing data/TN/TN_info_reduced4.csv')
lab=pd.read_csv('Processing data/TN/TN_researchlabs.csv')
```

```python
info.shape
```

```python
# Select and keep only the desired columns

print(lab.shape)

lab = lab[['MaskID', 'Date_of_Draw', 'Test_Name', 'Result']]

print(lab.shape)
```

```python
lab = lab.rename(columns={'Date_of_Draw': 'Visit_Dt'})
```

```python
#add column according to Test_Name

print(lab.shape)

unique_values = lab['Test_Name'].unique()

# Create new columns for each unique value
for value in unique_values:
    new_column_name = str(value)
    lab[new_column_name] = 0  # Initialize the new columns with None

# Display the updated DataFrame with new columns
print(lab.shape)
```

```python
df=lab
```

```python
#remove all typo result value (not float)

def is_float(value):
    try:
        float(value)
        return True
    except ValueError:
        return False

# Use the apply method to check each value in 'column1' and drop rows where the value cannot be converted to a float
lab = lab[lab['Result'].apply(is_float)]
print(lab.shape)
```

```python
exclude_columns = ['MaskID', 'Visit_Dt', 'Test_Name']

# Convert all columns to float except the excluded ones
for column in lab.columns:
    if column not in exclude_columns:
        lab[column] = lab[column].astype(float)

# Display the updated DataFrame
print(lab.shape)
```

```python
# embed the result 
df=lab

# Determine the total number of rows
total_rows = len(df)

# Define the progress interval (e.g., every 10%)
progress_interval = total_rows // 50  # 10% intervals

# Iterate through rows and print progress percentage
for index, row in df.iterrows():
    column_name = row['Test_Name']
    df.at[index, column_name] = row['Result']
    
    if index % progress_interval == 0:
        progress_percentage = (index / total_rows) * 100
        print(f"Progress: {progress_percentage:.2f}%")

# Drop the 'Result' column if you no longer need it
df = df.drop(columns=['Result'])

# Display the updated DataFrame
df.head(10)
```
